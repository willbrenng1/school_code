```python
#Riemann Sums Squares
n=100000
cum_sum =0
a=0
b=1
for i in range(1,100001):
   cum_sum+=sin(((a+((i*(b-a))/n))*pi)/2)


((b-a)/n)*cum_sum
```

```python
#Montee Carlor

n=1000
cum_sum =0
a=0
b=1
for i in range(1,1001):
   u= np.random.uniform(0, 1, 1)
   cum_sum+= sin(pi*(a+(b-a)*u)/2)


(1/n)*cum_sum
```
