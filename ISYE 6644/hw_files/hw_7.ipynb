```python

# Step 1: Generate 5000 samples of U and V ~ Uniform(0,1)
n_samples = 5000
U = np.random.uniform(0, 1, n_samples)
V = np.random.uniform(0, 1, n_samples)

# Step 2: Compute X = U + V
X = U + V

# Step 3: Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(X, bins=50, density=True, alpha=0.6, color='b', edgecolor='black')

# Overlay the theoretical triangular PDF
x_vals = np.linspace(0, 2, 1000)
pdf_vals = np.where(x_vals < 1, x_vals, 2 - x_vals)  # Triangular distribution

plt.plot(x_vals, pdf_vals, 'r-', linewidth=2, label="Theoretical PDF")
plt.xlabel("X = U + V")
plt.ylabel("Density")
plt.title("Histogram of X = U + V (5000 samples)")
plt.legend()
plt.show()
```
