```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn import tree
import numpy as np

from sklearn.svm import OneClassSVM

from sklearn.metrics import confusion_matrix,accuracy_score

from sklearn.model_selection import RandomizedSearchCV, train_test_split
```

```python
data = pd.read_csv('data/spambase.data', header=None).fillna(0)
```

```python
X=data.iloc[:,:-1]
y= data.iloc[:,-1]
```

3.1

```python
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X, y)
```

```python
plt.figure(figsize=(30,10))
tree.plot_tree(clf, feature_names=[str(e) for e in X.columns], class_names=[str(e) for e in y.unique()], filled=True,max_depth=3,fontsize=11)
plt.show()
```

3.2

```python
x_train, x_test,y_train,y_test = train_test_split(X,y, test_size=.25,random_state= 6740,shuffle=True)

```

```python
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(x_train, y_train)
```

```python
preds = clf.predict(x_test)

cart_misclass_rate = 1-accuracy_score(y_test,preds)
```

```python
n_trees = np.arange(10, 520, 20)
```

```python
misclassification_rates =[]
for t in n_trees:
    rf = RandomForestClassifier(n_estimators=t,random_state=6740)
    rf.fit(x_train, y_train)
    preds = rf.predict(x_test)
    misclassification_rates.append(1-accuracy_score(y_test,preds))
    
```

```python
np.argmin(misclassification_rates)
```

```python
n_trees[2]
```

```python
misclassification_rates[8]
```

```python
plt.plot(n_trees,misclassification_rates,label='Random Forest',color='g')
plt.axhline(y=cart_misclass_rate,color='r',label='CART')
plt.xlabel('Number of Trees')
plt.ylabel('Test Error - Misclassification Rate')
plt.title("Random Forest V. CART Misclassification Rate")
plt.legend()
```

3.3

```python
total_features = x_train.shape[1]
features=np.arange(1, total_features+1, 1)
misclassification_rates =[]
oob_errors = []
for f in features:
    rf = RandomForestClassifier(n_estimators=50,max_features=f,oob_score=True,random_state=6740)
    rf.fit(x_train, y_train)
    preds = rf.predict(x_test)
    misclassification_rates.append(1-accuracy_score(y_test,preds))
    oob_errors.append(1-rf.oob_score_)
```

```python
np.argmin(misclassification_rates),np.argmin(oob_errors)
```

```python
plt.plot(features,misclassification_rates,label='Misclassification Test',color='g')
plt.plot(features,oob_errors,label='OOB',color='r')

plt.xlabel('Number of Features')
plt.ylabel('Misclassification Rate - OOB Error')
plt.title("Test Error V. OOB Error Rates")
plt.legend()
```

3.4

```python
x_train_not_spam = x_train[y_train==0]
```

```python
nus = [.001,.01,.1,1/x_train_not_spam.shape[1],1/(x_train_not_spam.shape[1]*x_train_not_spam.var().sum())]
```

```python
gs=np.linspace(0,1)
```

```python
best_misclass , best_tune = float('inf'),None
mis_class_svm = []
params = []
for n in nus:
    for g in gs:
        svm_model_rbf = OneClassSVM(kernel='rbf',gamma=g,nu=n)
        svm_model_rbf.fit(x_train_not_spam)
        rbf_preds = svm_model_rbf.predict(x_test)
        rbf_preds=np.where(rbf_preds==1,0,1)
        mis_class =1-accuracy_score(y_test,rbf_preds)
        mis_class_svm.append(mis_class)
        params.append((n,g))
        if mis_class<best_misclass:
            best_misclass,best_tune = mis_class,(n,g)


```

```python
print(f'best nu: {best_tune[0]}')
print(f'best gamma: {best_tune[1]}')
print(f'error rate: {best_misclass}')
```
