```python
import numpy as np
# Parameters
m = 10
n = 2 # not used, but kept for consistency
# Random vector
x = np.random.randn(m, 1)
# Construct matrix X
X = np.vstack([1.5 * x.T, x.T])
# Compute covariance-like matrix
C = X @ X.T
# Eigen decomposition
eigvals, eigvecs = np.linalg.eig(C)
print("Eigenvalues:\n", eigvals)
print("Eigenvectors:\n", eigvecs)
```
