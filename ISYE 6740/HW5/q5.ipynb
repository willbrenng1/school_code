```python
f_0 = rng.normal(loc=0,scale=1,size=100)
f_1 = rng.normal(loc=0,scale=1.18,size=50)

test_data = np.concatenate([f_0,f_1])
```

```python
log_liklihood = np.log(norm.pdf(test_data,0,1.18)/norm.pdf(test_data,0,1))

wts = np.zeros(150)
```

```python
for x in range(1, 150):
    wts[x]=max(wts[x-1]+log_liklihood[x],0)
```

```python
plt.plot(wts) 
plt.axvline(x=100, color='r', linestyle='--', label='Actual Change')
plt.axvline(x=104, color='g', linestyle='--', label='Possible Change Detection')
plt.xlabel('time')                                                                  
plt.ylabel('CUSUM')                                                    
plt.legend()                 
plt.show()    

```
