```r
library(dplyr)
```

```r
data <- read.csv("KAG_wrangled_dataset.csv",stringsAsFactors = FALSE)
head(data) 
nrow(data) 
length(unique(data$ad_id))
```

```r
over_10000_impressions = filter(data,Impressions>10000)
```

```r
over_10000_impressions %>% group_by(gender) %>% summarise(percent=100*n()/nrow(over_10000_impressions))
```

```r
xyz_ads = filter(data,xyz_campaign_id==916)
xyz_ads$conversion_rate = xyz_ads$Approved_Conversion/xyz_ads$Total_Conversion
mean(xyz_ads$conversion_rate)
```

```r
data$cpc = data$Spent / data$Clicks
```

```r
clean_cpc_data = data[!is.na(data$cpc),]
```

```r
clean_cpc_data
```

```r
clean_cpc_data %>%
  group_by(gender, age) %>%
  summarise(mean_cpc = mean(cpc, na.rm = TRUE))
```

```r
data %>%
  group_by(gender, age) %>%
  summarise(mean_cpc = mean(cpc, na.rm = TRUE)) %>%
  arrange(desc(mean_cpc))
```
